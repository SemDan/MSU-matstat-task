# Дополнительное задание по МС

Импортируем pandas и считаем CSV файл

```python
import pandas as pd
```

```python
file = pd.read_csv("./storage/salaries.csv")
file.head(10)
```

## Найдем описательные статистики

- для признаков (где есть смысл)
- по каждой предложенной должности отдельно

### Salary in usd

```python
salaryUSD_col = file["salary_in_usd"]
salaryUSD_mean = salaryUSD_col.mean() # среднее
salaryUSD_median = salaryUSD_col.median()
salaryUSD_variance = salaryUSD_col.var()
salaryUSD_std = salaryUSD_col.std()

print("Salary in USD")
print("Mean: ", salaryUSD_mean)
print("Median: ", salaryUSD_median)
print("Variance: ", salaryUSD_variance)
print("Std: ", salaryUSD_std)
```

### Data Scientist

```python
DS_salaryUSD_col = file.loc[file["job_title"] == "Data Scientist"]["salary_in_usd"]

DS_salaryUSD_mean = DS_salaryUSD_col.mean() # среднее
DS_salaryUSD_median = DS_salaryUSD_col.median()
DS_salaryUSD_variance = DS_salaryUSD_col.var()
DS_salaryUSD_std = DS_salaryUSD_col.std()

print("Data Scientist salary in USD")
print("DS Mean: ", DS_salaryUSD_mean)
print("DS Median: ", DS_salaryUSD_median)
print("DS Variance: ", DS_salaryUSD_variance)
print("DS Std: ", DS_salaryUSD_std)
```

### Data Engineer

```python
DE_salaryUSD_col = file.loc[file["job_title"] == "Data Engineer"]["salary_in_usd"]

DE_salaryUSD_mean = DE_salaryUSD_col.mean() # среднее
DE_salaryUSD_median = DE_salaryUSD_col.median()
DE_salaryUSD_variance = DE_salaryUSD_col.var()
DE_salaryUSD_std = DE_salaryUSD_col.std()

print("Data Engineer salary in USD")
print("DE Mean: ", DE_salaryUSD_mean)
print("DE Median: ", DE_salaryUSD_median)
print("DE Variance: ", DE_salaryUSD_variance)
print("DE Std: ", DE_salaryUSD_std)
```

### Data Analyst

```python
DA_salaryUSD_col = file.loc[file["job_title"] == "Data Analyst"]["salary_in_usd"]

DA_salaryUSD_mean = DA_salaryUSD_col.mean() # среднее
DA_salaryUSD_median = DA_salaryUSD_col.median()
DA_salaryUSD_variance = DA_salaryUSD_col.var()
DA_salaryUSD_std = DA_salaryUSD_col.std()

print("Data Analyst salary in USD")
print("DA Mean: ", DA_salaryUSD_mean)
print("DA Median: ", DA_salaryUSD_median)
print("DA Variance: ", DA_salaryUSD_variance)
print("DA Std: ", DA_salaryUSD_std)
```
